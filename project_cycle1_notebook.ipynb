## 5. Sampling Simulation (Concept Preparation)

Goal: understand sampling variability of the sample mean.

TO DO:
1) Sample n = 30 observations (with replacement)  
2) Compute the sample mean of Glucose  
3) Repeat B = 1000 times  
4) Plot the distribution of sample means (matplotlib)  
5) Report the mean and standard deviation of the sampling distribution  


```python
# TO DO: set simulation parameters
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load dataset again (avoid df not defined error)
df = pd.read_csv("Diabetes.csv")

# simulation parameters
B = 1000
n = 30
```

```python
import numpy as np

means = []

for b in range(B):
    # TO DO: draw a bootstrap sample of size n (with replacement)
    sample = df.sample(n=n, replace=True)
    
    # TO DO: compute mean of Glucose in this sample
    m = sample["Glucose"].mean()
    
    means.append(m)

means = np.array(means)
```

```python
import matplotlib.pyplot as plt

plt.figure()

# TO DO: histogram of simulated sample means
plt.hist(means, bins=30)

plt.title("Sampling Distribution of Mean(Glucose), n=30")
plt.xlabel("Sample mean of Glucose")
plt.ylabel("Count")
plt.show()
```

```python
# TO DO: report center and spread of the sampling distribution
# hint: means.mean() and means.std(ddof=1)
center = means.mean()
spread = means.std(ddof=1)
center, spread
```

Interpretation (5–8 lines):
- What does the sampling distribution represent?
- Is it narrower than the raw Glucose distribution? Why?
- How does this connect to uncertainty of an estimate?


The sampling distribution represents the distribution of the sample mean of Glucose when we repeatedly take random samples of size 30 from the dataset. It shows how the sample mean varies due to random sampling variability.

From the histogram, the sampling distribution appears approximately bell-shaped and centered around 120.8. This center is close to the overall average Glucose level in the dataset.

Compared to the original Glucose distribution, the sampling distribution is much narrower. This happens because taking the average of a sample reduces variability.

The standard deviation of the sampling distribution (about 5.73) represents the uncertainty of the estimated sample mean. It tells us how much the sample mean may vary from sample to sample.

This demonstrates that while different samples may produce slightly different mean values, they tend to stay close to the true population mean, especially when the sample size is reasonably large.
